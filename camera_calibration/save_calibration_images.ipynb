```python
import numpy as np
import os
import cv2
import time
```

```python
# file paths
root_dir = "Z:/Isabel/arena/camera_calibration_files/"
img_date = input("input date of video acquisition (YYMMDD): ")

folder = f"{img_date}_calibration_videos"
```

```python
# file paths
root_dir = "Z:/Isabel/data/hpc_implants/PRL73/"
folder = "PRL73_240801"
```

```python
cam_ids = ['red_cam', 'yellow_cam', 'green_cam', 'blue_cam']
# cam_ids = ['rBack', 'rFront', 'rTop', 'lBack', 'lFront', 'lTop'] # SC cam IDs

n_video_folders = 1

# input start/end frames for each video
start_frames = np.asarray([1200*50]) # seconds * 50Hz
end_frames = np.asarray([1205*50])
```

```python
for i in range(n_video_folders):
    video_id = i+1
    
    print(f"\n{root_dir}{folder}_{video_id}")
    start_frame = start_frames[i]
    end_frame = end_frames[i]
    for cam_id in cam_ids:
        # make the image directory
        save_folder = f"{cam_id}_images/"
        save_path = f"{root_dir}{folder}/{save_folder}"
        if os.path.exists(save_path):
            print("warning, save folder already exists!")
        else:
            os.mkdir(save_path)

        # grab the images from the video file
        vid_file = f"{cam_id}.avi"
        tic = time.perf_counter()
        print(f"saving images from {cam_id} video...")
        cap = cv2.VideoCapture(f"{root_dir}{folder}/{vid_file}")
        for f in range(end_frame):
            ret, img = cap.read()
            if f < start_frame:
                continue
            else:
                if i == 0:
                    k = f
                else:
                    k = f + end_frames[i-1] # to add on images from other video
                f_zero_pad = str(k).zfill(len(str(end_frame))) # so images will be sorted correctly
                cv2.imwrite(f"{save_path}{cam_id}_{f_zero_pad}.tiff", img)
        cap.release()
        toc = time.perf_counter()
        print(f"{(toc - tic)/60:0.4f} minutes")
```
